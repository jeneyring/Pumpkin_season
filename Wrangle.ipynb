## Changing d-types:

```python
#first tackling date:
pumpkin_main['Date']= pd.to_datetime(pumpkin_main['Date'])
```

```python
#checking if true >>>>
pumpkin_main.info()
```

```python
#sorting by date:
pumpkin_main = pumpkin_main.sort_values(by='Date')
```

```python
#dropping Environment column:
pumpkin_main = pumpkin_main.drop(columns=['Environment'])
```

```python
#changing `Number of Stores` to int:
#first have to drop any commas:
pumpkin_main['Number of Stores']= pumpkin_main['Number of Stores'].apply(lambda x : x.replace(',',''))
```

```python
#now make Number of Stores into int:
pumpkin_main['Number of Stores']= pumpkin_main['Number of Stores'].astype(int)
```
